```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import percentileofscore

# Assuming you have already loaded your DataFrame data

# Replace "unavailable" with NaN in 'Rating' and 'Revenue (Millions)'
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')
data['Revenue (Millions)'] = pd.to_numeric(data['Revenue (Millions)'], errors='coerce')

# Scatter plot
sns.scatterplot(x='Rating', y='Revenue (Millions)', data=data)
plt.title('Scatter Plot of Rating vs Revenue')
plt.xlabel('Rating')
plt.ylabel('Revenue (Millions)')
plt.show()

# Calculate the percentile for Rating and Revenue
rating_percentile = percentileofscore(data['Rating'], data['Rating'].iloc[0])
revenue_percentile = percentileofscore(data['Revenue (Millions)'], data['Revenue (Millions)'].iloc[0])

print(f'Rating Percentile: {rating_percentile:.2f}%')
print(f'Revenue Percentile: {revenue_percentile:.2f}%')

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Assuming you have already loaded your DataFrame df

# Selecting the columns for box plot
df_box_plot = data[['Revenue (Millions)', 'Metascore', 'Year', 'Rating', 'Runtime (Minutes)', 'Votes']]

# Create a box plot using Seaborn
plt.figure(figsize=(18, 10))
sns.set(style="whitegrid")
i = 1

outlier_percentage = {}  # Dictionary to store outlier percentages

for feature in df_box_plot.columns:
    plt.subplot(3, 3, i)
    sns.boxplot(x=df_box_plot[feature], y=None, palette="rocket_r", legend=False)  # Assign y=None and set legend=False

    # Calculate the first quartile (Q1) and third quartile (Q3)
    q1 = df_box_plot[feature].quantile(0.25)
    q3 = df_box_plot[feature].quantile(0.75)

    # Calculate the interquartile range (IQR)
    iqr = q3 - q1

    # Define the lower and upper bounds for outliers
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Count the number of outliers
    outliers = df_box_plot[(df_box_plot[feature] < lower_bound) | (df_box_plot[feature] > upper_bound)]
    outlier_percentage[feature] = len(outliers) / len(df_box_plot) * 100

    i += 1

plt.show()

# Display the percentage of outliers for each feature
print("Percentage of Outliers:")
for feature, percentage in outlier_percentage.items():
    print(f"{feature}: {percentage:.2f}%")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
# If not, load your data into df

# Select features and prediction label
features = data[['Year', 'Runtime (Minutes)', 'Votes', 'Revenue (Millions)', 'Metascore']]
prediction_label = data['Rating']

# Concatenate features and prediction label
data_for_correlation = pd.concat([features, prediction_label], axis=1)

# Calculate correlation matrix
correlation_matrix = data_for_correlation.corr()

# Visualize correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation with Rating')
plt.show()

```
